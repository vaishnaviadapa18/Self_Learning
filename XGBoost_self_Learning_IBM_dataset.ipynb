```python
clf_xgb = xgb.XGBClassifier(objective = 'binary:logistic', 
                            missing = 0, 
                            seed = 42,
                            base_score = 0.5,
                            gamma=0,
                            learning_rate = 0.1,
                            booster = 'gbtree',
                            min_child_weight = 1,
                           max_depth = 3,
                           random_state = 0,
                           verbosity = 1,
                           reg_lambda = 1)
```

```python
clf_xgb.set_params(early_stopping_rounds=10, eval_metric='aucpr')
```

```python
clf_xgb_check = clf_xgb.fit(X_train,
            y_train,
            verbose = True,
            #early_stopping_rounds= 10,
            #eval_metric='aucpr',
            eval_set = [(X_test,y_test)]) 
#early stopping. We are building trees. At some point, prediction will not improve. Here 10 means, 10 more trees are built and if none of them add value, the iterations stop 

# using AUC to evaluate how well the predictions are made. We pass in the testing data set as well. We train the trees on training dataset.
#It is evaluating how many trees to create and when to stop by using the testing dataset 
```

```python
#Hyper parameter tuning 
#This data is imbalanced. Which is only 27% are churned customers. So, we use another attribute called 
#scale_pos_weight -  you can start setting scale_pos_weight = count_negative_samples / count_positive_samples
# because the model trains well on majority data, but may not train well on minority data. So, to balance that imbalance,
#we try to give more weightage to the minority groups when creating subsequent trees.
```

```python
#Round 1 
param_grid = {'max_depth' : [3,4,5],
              'learning_rate' : [0.1,0.01,0.05],
              'gamma' : [0,1,2],
              'reg_lambda' : [0,1,10],
              'scale_pos_weight': [1,3,5]
              }
```

```python
param_grid
```

```python
optimal_params = GridSearchCV(estimator=xgb.XGBClassifier(objective= 'binary:logistic',
                             seed = 42,
                             sub_sample = 0.9,
                             colsample_bytree = 0.5),
                              param_grid = param_grid,
                              scoring = 'roc_auc',
                              verbose = 2,
                              n_jobs = 10,
                              cv = 3)
```

```python
optimal_params.fit(X_train, 
                   y_train, 
                   early_stopping_rounds=10, 
                   eval_metric='auc',
                   verbose = True,
                  eval_set = [(X_test,y_test)])
```

```python
param_grid = {'max_depth' : [10],
              'learning_rate' : [0.05],
              'gamma' : [2],
              'reg_lambda' : [100,150],
              'scale_pos_weight': [1,3,5]
              }
```
